# Scraping

```python
# Scraping web google play store for app reviews and ratings with google_play_scraper library dan app store scraper library

from app_store_scraper import AppStore
from google_play_scraper import app, reviews_all, Sort
import pandas as pd
import time
```

# Scraping google play store

```python
# Define the app id dana dompet digital
app_id_dana = 'id.dana'

# Get the app details
result = reviews_all(
    app_id=app_id_dana,
    lang='id',
    count=1000,
    sort=Sort.NEWEST,
    filter_score_with=None
)
```

```python
# Create a dataframe
data_google_play = pd.DataFrame(result)
```

```python
data_google_play.tail()
```

```python
# Save the data to csv
data_google_play.to_csv('Dataset/data_google_play.csv', index=False)
```

# Scraping data app store

```python

# Define the app id dana dompet digital
app_id_dana_appstore = 1437123008

# Get the app details
g_app = reviews_all(
    app_id=app_id_dana_appstore,
    sleep_milliseconds=0,
    lang='id',
    sort=Sort.NEWEST
)

a_reviews = AppStore(country="id", app_name="dana-dompet-digital", app_id=app_id_dana_appstore)
a_reviews.review(how_many=1000) 
```
